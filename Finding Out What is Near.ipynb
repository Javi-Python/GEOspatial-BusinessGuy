```python
from pymongo import MongoClient
import pandas as pd
import requests
import json
from dotenv import load_dotenv
import os
import pandas as pd
load_dotenv()
import api_calls as api
```

```python
#We connect to our mongo database collection with the venues we extracted from 4square
client = MongoClient('localhost:27017')
db = client.get_database("ironhack")
geo = db.get_collection("Geobusinesses")
```

```python
from pymongo import GEOSPHERE
geo.create_index([("geometry", GEOSPHERE)])
```

```python
geo.find_one({})
```

```python
# We choose a companies coordinates for every city in our analysis.

paypal_sanjose = [-121.9276, 37.3209]
paypal_sanjose_point = {"type": "Point",  "coordinates": paypal_sanjose}

google_mtview = [-122.084143, 37.421972]
google_mtview_point = {"type": "Point",  "coordinates": google_mtview}

yahoo_sunny = [-122.025485, 37.418531]
yahoo_sunny_point = {"type": "Point",  "coordinates": yahoo_sunny}

vmware_palo = [-122.145981, 37.403315] 
vmware_palo_point = {"type": "Point",  "coordinates": vmware_palo}

twitter_sanfran = [-122.416924, 37.776805]
twitter_sanfran_point = {"type": "Point",  "coordinates": twitter_sanfran}
```

```python
#For us to consider something near, we will use a radius of 3kms, we employ the function near point, 
#that checks what is near each business location within our chosen radius
radius = 3000
near_paypal = api.near_point(paypal_sanjose_point, radius)
near_google = api.near_point(google_mtview_point, radius)
near_yahoo = api.near_point(yahoo_sunny_point, radius)
near_vmware = api.near_point(vmware_palo_point, radius)
near_twitter = api.near_point(twitter_sanfran_point, radius)
```

```python
near_paypaldf = pd.DataFrame(near_paypal)
near_googledf = pd.DataFrame(near_google)
near_yahoodf = pd.DataFrame(near_yahoo)
near_vmwaredf = pd.DataFrame(near_vmware)
near_twitterdf = pd.DataFrame(near_twitter)
```

```python
#As we can see there are a lot of duplicate values.
near_googledf
```

```python
#We create a column that specifies if the value is duplicated by assigning True or False if it is.
near_googledf['duplicated'] = near_googledf.location.astype(str).duplicated(keep= 'first')
near_paypaldf['duplicated'] = near_paypaldf.location.astype(str).duplicated(keep= 'first')
near_yahoodf['duplicated'] = near_yahoodf.location.astype(str).duplicated(keep= 'first')
near_vmwaredf['duplicated'] = near_vmwaredf.location.astype(str).duplicated(keep= 'first')
near_twitterdf['duplicated'] = near_twitterdf.location.astype(str).duplicated(keep= 'first')
```

```python
#We create a filter that picks up only the non duplicated venues, or the first value of the duplicated columns.
near_paypal_filtered = near_paypaldf[near_paypaldf['duplicated'] == False]
near_google_filtered = near_googledf[near_googledf['duplicated'] == False]
near_yahoo_filtered = near_yahoodf[near_yahoodf['duplicated'] == False]
near_vmwared_filtered = near_vmwaredf[near_vmwaredf['duplicated'] == False]
near_twitter_filtered = near_twitterdf[near_twitterdf['duplicated'] == False]
```

```python
#We count the total values in our filtered dataframes and print them out
print(f'paypal has a total of: {len(near_paypal_filtered)} venues we are interested in')
print(f'google has a total of: {len(near_google_filtered)} venues we are interested in')
print(f'yahoo has a total of: {len(near_yahoo_filtered)} venues we are interested in')
print(f'vmware has a total of: {len(near_vmwared_filtered)} venues we are interested in')
print(f'twitter has a total of: {len(near_twitter_filtered)} venues we are interested in')
```
