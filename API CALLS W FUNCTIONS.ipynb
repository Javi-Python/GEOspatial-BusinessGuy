# Extracting NightClubs

```python
#We extract nightclub information for all the cities in our analysis
sanjose_night = api.call_4square(sanjose, 'nightclubs')
mtview_night = api.call_4square(mtview, 'nightclubs')
sunnyvale_night = api.call_4square(sunnyvale, 'nightclubs')
paloalto_night = api.call_4square(paloalto, 'nightclubs')
sanfrancisco_night = api.call_4square(sanfrancisco, 'nightclubs')
```

```python
sanfrancisco_night
```

```python
#We dive into the part of the response we are interested in using our get_response function.
sanj_r = api.get_response(sanjose_night)
mtv_r = api.get_response(mtview_night)
sunny_r = api.get_response(sunnyvale_night)
palo_r = api.get_response(paloalto_night)
sanfran_r = api.get_response(sanfrancisco_night)
```

```python
sanj_r
```

```python
#We create a list of dictionaries with the name latitude longitude and type of business
sanjose2 = api.name_lat_lon(sanj_r, 'nightclubs')
mtview2 = api.name_lat_lon(mtv_r, 'nightclubs')
sunny2 = api.name_lat_lon(sunny_r, 'nightclubs')
palo2 = api.name_lat_lon(palo_r, 'nightclubs')
sanfran2 = api.name_lat_lon(sanfran_r, 'nightclubs')
```

```python
sanjose3 = api.to_point(sanjose2)
mtview3 = api.to_point(mtview2)
sunny3 = api.to_point(sunny2)
paloalto3 = api.to_point(palo2)
sanfran3 = api.to_point(sanfran2)
```

```python
sanjose3 = pd.DataFrame(sanjose3)
mtview3 = pd.DataFrame(mtview3)
sunny3 = pd.DataFrame(sunny3)
paloalto3 = pd.DataFrame(paloalto3)
sanfran3 = pd.DataFrame(sanfran3)
paloalto3
```

```python
sanjose3.to_json('nigthclubs_san_jose5.json', orient = 'records')
mtview3.to_json('nightclubs_mtvie2.json', orient = 'records')
sunny3.to_json('nightclubs_sunny2.json', orient = 'records')
paloalto3.to_json('nightclubs_palo2.json', orient = 'records')
sanfran3.to_json('nightclubs_sanfran.json', orient = 'records')
```
