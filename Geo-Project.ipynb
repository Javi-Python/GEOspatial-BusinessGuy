We create a projection with the data that we are interested to filter companies

For our first filter of the databases in the company, we decided to go with companies that have raised over 1M in funds, and have more than 1000 employees. We are interested in a city that has big companies that we can do partnerships with as well as a powerfull financial powerhouse to raise funding.

```python
proj = {'name':1, 'offices.city' : 1, 'category_code': 1, 'total_money_raised':1, 'offices.country_code':1, 'number_of_employees':1, 'offices.latitude':1, 
        'offices.longitude': 1, 'offices.state_code': 1}

million = {'total_money_raised': {'$regex': "M"}}
billion = {'total_money_raised': {'$regex': "B"}}
both_mb = {'$or' : [million, billion]}
filtro2 = {'number_of_employees': {'$gte': 1000}}

lista_1m = list(comp.find({'$and': [both_mb, filtro2]}, proj))
```

```python
#We create a list of Parameters for the 4 Square Api
#We will start by exploring the party scene of our selected cities. 
params_sanjose_nightclub = {
    'client_id': tok1,
    'client_secret': tok2,
    'v' : "20200323",
    'll' : f"{sanjose.get('coordinates')[0]}, {sanjose.get('coordinates')[1]}",
    'query' : 'nightclub'
    }

params_mtview_nightclub = {
    'client_id': tok1,
    'client_secret': tok2,
    'v' : "20200323",
    'll' : f"{mtview.get('coordinates')[0]}, {mtview.get('coordinates')[1]}",
    'query' : 'nightclub'
    }
params_sunnyvale_nightclub = {
    'client_id': tok1,
    'client_secret': tok2,
    'v' : "20200323",
    'll' : f"{sunnyvale.get('coordinates')[0]}, {sunnyvale.get('coordinates')[1]}",
    'query' : 'nightclub'
    }
params_paloalto_nightclub = {
    'client_id': tok1,
    'client_secret': tok2,
    'v' : "20200323",
    'll' : f"{paloalto.get('coordinates')[0]}, {paloalto.get('coordinates')[1]}",
    'query' : 'nightclub'
    }
params_sanfrancisco_nightclub = {
    'client_id': tok1,
    'client_secret': tok2,
    'v' : "20200323",
    'll' : f"{sanfrancisco.get('coordinates')[0]}, {sanfrancisco.get('coordinates')[1]}",
    'query' : 'nightclub'
    }
```

# We look for Night Life in San Jose

```python
sanjose_night = requests.get(url= url_query, params = params_sanjose_nightclub).json()
```

```python
sanjose_night
```

```python
sj_clubs = sanjose_night.get("response").get("groups")[0].get("items")
sj_clubs
```

```python
from functools import reduce
import operator
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
mapa_nombre = ['venue', 'name']
mapa_lat = ['venue', 'location', 'lat']
mapa_long = ['venue', 'location', 'lng']
```

```python
lista_sj_clubs = []
for dicc in sj_clubs:
    lista = {}
    lista["name"] = getFromDict(dicc, mapa_nombre)
    lista["latitud"]= getFromDict(dicc, mapa_lat)
    lista["longitud"] = getFromDict(dicc,mapa_long)
    lista_sj_clubs.append(lista)
```
